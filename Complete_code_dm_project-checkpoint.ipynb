```python
import networkx as nz
G_symmetric = nz.Graph()
G_fb = nz.read_edgelist('facebook_combined.txt', create_using = nz.Graph(), nodetype=str)

print('press 1 to get number of nodes')
print('press 2 to get number of edges')
print('press 3 to get number of connected nodes')
print('press 4 to draw graph')
print('press 5 to get  path')
print('press 6 to least connected person')
print('press 7 to get information')
print('press 8 to get triadic closure')
print('press 9 to get information')
print('press 10 to get most connected person')
print('press 11 to get mutual connection')
print('press 12 to add edge')
print('press 13 to add node')

ch=input('enter your choice')
if ch=='1':
    print(G_fb.number_of_nodes()) 
    print(G_fb.nodes())  
if ch=='2':
    print(G_fb.number_of_edges())  
    print(G_fb.edges()) 
if ch=='3':
    count=0
    nodee=input('enter the node')
    for n in G_fb.neighbors(nodee):
        print(n)
        count+=1
    print("no.of edges=",count)
if ch=='4':
    nz.draw(G_fb)  

if ch=='5': 
    s=input("source node=?")
    d=input("destination node=?")
    print(nz.shortest_path(G_fb,source=s,target=d))
if ch=='6': 
       
    li=[]

    minnode=0
    minn=10
    for i in G_fb:
        
        li=list(nz.neighbors(G_fb,i))

        if(minn>len(li)):
            minn=len(li)
            minnode=i 
    print("PERSON= "+str(minnode)+"  connections= "+str(minn))
        
    
    
if ch=='7':
    print(nz.info(G_fb))
if ch=='8':
    triadic_closure = nz.transitivity(G_fb)
    print("Triadic closure:", triadic_closure)
if ch=='9': 
    print(list(nz.nodes_with_selfloops(G_fb)))
if ch=='10': 
    li=[]
    maxx=0
    mn=0
    minnode=0
    minn=10
    for i in G_fb:
        li=list(nz.neighbors(G_fb,i))
        if(maxx<len(li)):
            maxx=len(li)
            mn=i
        if(minn>len(li)):
            minn=len(li)
            minnode=i
    print("PERSON= "+str(mn)+"  connections= "+str(maxx))
    
    
if ch=='11':
    p1=input("person A?")
    p2=input("person B?")
    print('MUTUAL CONNECTIONS')
  
    
    print(sorted(nz.common_neighbors(G_fb, p1,p2)))
if ch=='12':
    
    person1=input("person 1")
    person2=input("person 2")
    
  
    G_fb.add_edge(person1,person2)
    print('CONNECTION ADDED SUCCESSFULLY')
if ch=='13':
    
    person1=input("enter person")
    
    
  
    G_fb.add_node(person1)
    print('PERSON ADDED SUCCESSFULLY')
```
